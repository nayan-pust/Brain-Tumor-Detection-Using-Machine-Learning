```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#POLAR NRZ
N=10
n=np.random.randint(2,size=N)
print(n)
nn=np.where(n==1,1,-1)
t=np.arange(0,N+0.01,0.01)
y=np.zeros(len(t))
for i in range(N):
    y[i * 100:(i + 1) * 100] = nn[i]
    
plt.plot(t,y,linewidth=2)
plt.grid(True)
plt.axis([0,N,-1.5,1.5])
plt.title("Polar NRZ") 
plt.show()   
```

```python
N=10
n=np.random.randint(2,size=N)
print(n)
i=1
t=np.arange(0,N+0.01,0.01)
y=np.zeros(len(t))
for j in range(len(t)):
    if t[j]<=i:
        y[j] = n[i-1]
    else:
        y[j]=n[i-1]
        i+=1
        
plt.plot(t,y,linewidth=2)
plt.grid(True)
plt.axis([0,N,-1.5,1.5])
plt.title("Unipolar NRZ") 
plt.show()          
```

```python
N = 10
n = np.random.randint(2, size=N)
print(n)
t = np.arange(0, N + 0.01, 0.01)
y = np.zeros(len(t))
for i in range(N):
    for j in range(len(t)):
        if t[j]>=i and t[j]<=i+0.5:
            y[j]= n[i]
        elif t[j] > i+0.5:
            break
        
plt.plot(t, y, linewidth=2)
plt.grid(True)
plt.axis([0, N, -1.5, 1.5])
plt.title("Unipolar Return to Zero")
plt.show()        
```

```python
N = 10  # Number of bits
n = np.random.randint(2, size=N)  # Random bit generation
print(n)
nnn = []
for bit in n:
    if bit == 1:
        nn = [1, -1]
    else:
        nn = [-1, 1]
    nnn.extend(nn)  
i = 0
l = 0.5
t = np.arange(0, N + 0.01, 0.01)
y = []
for j in range(len(t)):
    if t[j] <= l:  # Condition for the first half cycle
        y.append(nnn[i])  # Assign values for the first half cycle
    else:
        y.append(nnn[i])
        i += 1
        l += 0.5

plt.plot(t, y, linewidth=2)  # Linewidth 2 for clear visualization
plt.axis([0, N, -1.5, 1.5])  # Axis set-up
plt.grid(True)
plt.title("Manchester Coding")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([1,0,1,0,1,0,1,0,1])
bp=0.000001
print("Binary Information at Transmitter:")
print(x)

bit=np.array([np.ones(100) if bit ==1 else np.zeros(100) for bit in x]).flatten()
t1=np.arange(bp/100, 100 * len(x) * (bp/100) + bp/100 ,  bp/100)
plt.subplot(211)
plt.plot(t1,bit,linewidth=2.5)
plt.grid(True)
plt.axis([0,bp*len(x),-0.5,1.5])
plt.xlabel("Time(Sec)")
plt.ylabel("Amplitude(Volt)")
plt.title("Transmitting Information as Digital Signal")

A1,A2=10,5
br=1/bp
f=br*10
t2=np.arange(bp/99, bp + bp/99 ,  bp/99)
m=np.concatenate([(A1*np.cos(2*np.pi*f*t2)) if bit==1 else (A2*np.cos(2*np.pi*f*t2)) for bit in x])
t3=np.arange(bp/99, bp*len(x) ,  bp/99)
plt.subplot(212)
plt.plot(t3,m,linewidth=2.5)
plt.grid(True)
plt.axis([0,bp*len(x),-10.5,10.5])
plt.xlabel("Time(Sec)")
plt.ylabel("Amplitude(Volt)")
plt.title("Waveform for Binary ASK modulation corresponding Binary Information")
plt.tight_layout()
plt.show()
```
